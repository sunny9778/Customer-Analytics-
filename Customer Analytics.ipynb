```python
df = pd.read_excel('Data_Spreadsheet.xlsx')
```

```python
segment = df['Customer Segment'].value_counts()
colors = ("orange", "cyan", "brown", "grey")
explode = (0.2, 0.0, 0.0, 0.2)
plt.pie(segment, labels=segment.index, autopct = "%1.1f%%", startangle = 90, explode = explode, shadow = True, colors = colors)
plt.title("Fig.1 Customer Segment")
plt.legend(loc="upper left")
plt.show()
```

```python
category = df['Product Category'].value_counts()
plt.pie(category, labels=category.index, autopct = "%1.1f%%", startangle=90, shadow = True)
plt.title("Fig.2 Product Category")
plt.legend(loc="upper left")
plt.show()
```

```python
Container = df['Product Container'].value_counts()
labels=Container.index
plt.bar(labels, Container)
plt.title('Shipping of different Product Container')
plt.xlabel("Container")
plt.ylabel("Shipping Cost")
plt.show()
```

```python
Container = df['Product Container'].value_counts()
plt.pie(Container, labels=Container.index, autopct = "%1.0f%%", startangle=90, shadow = True)
plt.title("Fig.3 Product Container")
plt.show()
```

```python
plt.style.use("ggplot")
plt.xlabel("Profit")
plt.ylabel("Shipping Cost")
plt.title("Relationship between Profit & Shipping Cost")
plt.scatter(df["Profit"],df["Shipping Cost"], c = "m")
plt.show()
```

```python
ShipMode = df['Ship Mode'].value_counts()
plt.pie(ShipMode, labels=ShipMode.index, autopct = "%1.0f%%", startangle=90, shadow = True)
plt.title("Ship Mode")
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```
